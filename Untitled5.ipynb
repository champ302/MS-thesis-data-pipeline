```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Professional Styling for Thesis
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams.update({'font.size': 12, 'font.family': 'serif'})

# --- PATH SETUP (As per your Screenshots) ---
folder_path = r'E:\waheed\MS thesis\dataset'

# Files ke sahi naam screenshots ke mutabiq
file1 = os.path.join(folder_path, 'Final Experimental Data Tabl.xlsx')
file2 = os.path.join(folder_path, 'Model Comparison (Ours vs Baselines).xlsx')

# 1. Load Data (Using read_excel because files are .xlsx)
main_results = pd.read_excel(file1)
comparison_data = pd.read_excel(file2)

# Pre-processing (Taking first 5 workload rows)
main_results = main_results.head(5) 

# --- PLOT 1: THROUGHPUT STABILITY (Linear Scalability) ---
plt.figure(figsize=(10, 6))
plt.plot(main_results['Workload'], main_results['Throughput (Rec/s)'], 
         marker='o', linestyle='-', color='darkblue', linewidth=2.5, label='Actual Throughput')
plt.axhline(y=main_results['Throughput (Rec/s)'].mean(), color='red', linestyle='--', label='Mean Throughput')

plt.title('Figure 4.1: System Throughput Stability Analysis', fontweight='bold')
plt.xlabel('Workload (Number of Records)')
plt.ylabel('Throughput (Records per Second)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.savefig('throughput_plot.png', dpi=300)
plt.show()

# --- PLOT 2: COMPONENT LATENCY CONTRIBUTION (Stacked Bar) ---
plt.figure(figsize=(10, 6))
workloads = main_results['Workload'].astype(str)
vision = main_results['Ensemble Vision (s)']
text_cnn = main_results['Text CNN (s)']

plt.bar(workloads, vision, label='Vision Ensemble (ResNet50+VGG16)', color='#4c72b0')
plt.bar(workloads, text_cnn, bottom=vision, label='Text Engine (1D-CNN)', color='#dd8452')

plt.title('Figure 4.2: Cumulative Latency Contribution', fontweight='bold')
plt.xlabel('Workload Size')
plt.ylabel('Latency in Seconds')
plt.legend()
plt.savefig('latency_stacked.png', dpi=300)
plt.show()

# --- PLOT 3: MODEL COMPARISON (Baselines vs Ours) ---
plt.figure(figsize=(10, 6))
# Cleaning latency string '20.0s' to float 20.0
comparison_data['Latency_Value'] = comparison_data['Latency (for 1000 rec)'].str.replace('s', '', case=False).astype(float)

sns.barplot(x='Model', y='Latency_Value', data=comparison_data, palette='viridis')
plt.title('Figure 4.3: Latency Comparison - Proposed vs Baselines (1000 Records)', fontweight='bold')
plt.ylabel('Latency (Seconds)')
plt.xticks(rotation=15)
plt.savefig('comparison_bar.png', dpi=300)
plt.show()

print("Graphs successfully generated using data from E:\\waheed\\MS thesis\\dataset")
# import pandas as pd
# import matplotlib.pyplot as plt
# import seaborn as sns

# # Professional Styling for Thesis
# plt.style.use('seaborn-v0_8-whitegrid')
# plt.rcParams.update({'font.size': 12, 'font.family': 'serif'})

# # 1. Load Data from your approved CSVs
# main_results = pd.read_csv('Final Experimental Data Tabl.xlsx - Sheet1.csv')
# comparison_data = pd.read_csv('Model Comparison (Ours vs Baselines).xlsx - Sheet1.csv')

# # Pre-processing (Cleaning names if necessary)
# main_results = main_results.head(5) # Taking first 5 workload rows

# # --- PLOT 1: THROUGHPUT STABILITY (Linear Scalability) ---
# plt.figure(figsize=(10, 6))
# plt.plot(main_results['Workload'], main_results['Throughput (Rec/s)'], 
#          marker='o', linestyle='-', color='darkblue', linewidth=2.5, label='Actual Throughput')
# plt.axhline(y=main_results['Throughput (Rec/s)'].mean(), color='red', linestyle='--', label='Mean Throughput')

# plt.title('Figure 4.1: System Throughput Stability Analysis', fontweight='bold')
# plt.xlabel('Workload (Number of Records)')
# plt.ylabel('Throughput (Records per Second)')
# plt.legend()
# plt.grid(True, alpha=0.3)
# plt.savefig('throughput_plot.png', dpi=300)
# plt.show()

# # --- PLOT 2: COMPONENT LATENCY CONTRIBUTION (Stacked Bar) ---
# plt.figure(figsize=(10, 6))
# workloads = main_results['Workload'].astype(str)
# vision = main_results['Ensemble Vision (s)']
# text_cnn = main_results['Text CNN (s)']

# plt.bar(workloads, vision, label='Vision Ensemble (ResNet50+VGG16)', color='#4c72b0')
# plt.bar(workloads, text_cnn, bottom=vision, label='Text Engine (1D-CNN)', color='#dd8452')

# plt.title('Figure 4.2: Cumulative Latency Contribution', fontweight='bold')
# plt.xlabel('Workload Size')
# plt.ylabel('Latency in Seconds')
# plt.legend()
# plt.savefig('latency_stacked.png', dpi=300)
# plt.show()

# # --- PLOT 3: MODEL COMPARISON (Baselines vs Ours) ---
# plt.figure(figsize=(10, 6))
# # Cleaning latency string '20.0s' to float 20.0
# comparison_data['Latency_Value'] = comparison_data['Latency (for 1000 rec)'].str.replace('s', '').astype(float)

# sns.barplot(x='Model', y='Latency_Value', data=comparison_data, palette='viridis')
# plt.title('Figure 4.3: Latency Comparison - Proposed vs Baselines (1000 Records)', fontweight='bold')
# plt.ylabel('Latency (Seconds)')
# plt.xticks(rotation=15)
# plt.savefig('comparison_bar.png', dpi=300)
# plt.show()
```
